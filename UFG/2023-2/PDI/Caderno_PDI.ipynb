# Processamento Digital de Imagens

## Capítulo 1 - Introdução

* Interesse:

    * Melhora das informações visuais para a interpretação humana.
    * Processamento de dados de imagens para armazenamento, transmissão e representação, considerando a percepção automática por máquinas.

### O que é processamento digital de imagens?

* Uma imagem pode ser definidade como uma função bidimensional , $f(x, y)$, em que $x$ e $y$ são as coordenadas *espaciais* (plano), e a amplitude de $f$ em qualquer par de coordenadas $(x, y)$ é chamada de *intensidade* ou *nível de cinza* da imagem nesse ponto.

* Quando $x$, $y$ e os valores de intensidade de $f$ são quantidades **finitas** e **discretas**, chamamos de *imagem digital*. 

* *Processamento digital de imagens* se refere ao processo de  imagens digitais por uma computador digital.

* Uma imagem digital é composta de um número finito de elementos, cada uma com **localização** e **valor** específicos. Esses elementos são chamados de **pixels**.

* Olhos humanos exergam a banda visual do espectro eletromagético (EM). Já os aparelhos de processamento de imagens cobrem quase todo eo espectro EM, variando de ondas gama a ondas de rádio.
