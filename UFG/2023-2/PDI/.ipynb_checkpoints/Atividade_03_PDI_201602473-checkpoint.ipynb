## Questão 01

Compreenda a implementação do cálculo e exibição de um histograma tal como a contida no Jupyter Notebook fornecido abaixo. Implemente seu algoritmo com binning usando como referência o código do livro do Burger (veja último slide da aula acima).

```python
# importando libs
import cv2  #---------------------lib OpenCV
import numpy as np #--------------lib manipulão de arrays
import matplotlib.pyplot as plt #-lib para plotagem de gráficos
from PIL import Image
from matplotlib import image as mpimg
```

```python
# Função que retorna array para histograma binned
# ----Dei preferência para que o número de bins e valores de cinza fossem
# ----informados como argumentos
def binnedHistogram(img, numvalues, bins):
    h = np.zeros(bins) # Inicializa uma array só de zeros do tamanho do  num de bins
    width, height = img.shape

    for v in range(height):
        for u in range(width):
            a = img[u,v] # pega a intensidade de cinza do pixel (u,v)
            i = a * bins // numvalues # coloca no bin calculado forçando ser um inteiro
            h[i] = h[i] + 1
    
    return h
```
